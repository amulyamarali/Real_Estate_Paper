```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import mean_squared_error

# Step 1: Read the CSV File
df = pd.read_csv('final_clean_1.csv')

# Step 2: Preprocess the Data
# Handle missing values, encode categorical variables if necessary

# Assuming 'price' is the target variable
y = df['price']
X = df.drop(columns=['price'])

# Step 3: Vectorize Categorical Variables
# Assuming 'locality' is a categorical variable, use OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(X[['locality']])
X_encoded_df = pd.DataFrame(X_encoded, columns=encoder.get_feature_names_out(['locality']))

# Remove 'locality' from numerical_cols
numerical_cols = ['sqft', 'bedroom', 'bathroom', 'balcony', 'age', 'ownership', 'flooring', 'furnishing', 'facing', 'parking', 'gated_security', 'Children', 'Club', 'Fire', 'Gas', 'Gym', 'Intercom', 'Internet', 'Lift', 'Park', 'Security', 'Sewage', 'Shopping', 'Swimming', 'Visitor']

# Step 4: Scale numerical variables
scaler = StandardScaler()
X_numerical = scaler.fit_transform(X[numerical_cols])

# Step 5: Concatenate categorical and numerical features
X_processed = pd.concat([pd.DataFrame(X_numerical, columns=numerical_cols), X_encoded_df], axis=1)

# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.05, random_state=42)

# Step 4: Hyperparameter Tuning using Grid Search
param_grid = {
    'learning_rate': [0.05, 0.1, 0.15],
    'max_depth': [3, 4, 5],
    'min_child_weight': [1, 2, 3],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

xgb_model = xgb.XGBRegressor()
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_

# Step 5: XGBoost Model Training with Best Parameters
best_model = xgb.XGBRegressor(**best_params)
best_model.fit(X_train, y_train)

# Step 6: Model Evaluation
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Step 4: XGBoost Model Training
#model = xgb.XGBRegressor()
#model.fit(X_train, y_train)

# Step 5: Model Evaluation
score = best_model.score(X_test, y_test)
print("Model Score:", score)
```
