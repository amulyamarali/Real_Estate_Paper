```python
# read csv file 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read csv file
df = pd.read_csv('final.csv')
df.head()
```

```python
#print all the columns
print(df.columns)
```

```python
# remove 'web-scraper-order', 'web-scraper-start-url', 'link01', 'link01-href'
df = df.drop(['web-scraper-order', 'web-scraper-start-url', 'link01', 'link01-href'], axis=1)

df.head()
```

```python
# print the type of each column
print(df.dtypes)
```

```python
df['price'].unique()
```

```python
for i in df['price']:
    x = i.split(' ')
    if x[2] == 'Lacs':
        df['price'] = df['price'].replace(i, float(x[1])*100000)
    if x[2] == 'Crores':
        df['price'] = df['price'].replace(i, float(x[1])*10000000)
    if x[2] == 'Crore':
        df['price'] = df['price'].replace(i, float(x[1])*10000000)
```

```python
# convert the price column to float
df['price'] = df['price'].astype(float)
```

```python
# print type of price column
print(df['price'].dtype)
```

```python
df.head()
```

```python
# remove the comma from 'sqft' column values
df['sqft'] = df['sqft'].str.replace(',', '')

# convert 'sqft' column to integers
df['sqft'] = df['sqft'].astype(int)
```

```python
for i in df['bedroom']:
    x = i.split(' ')
    df['bedroom'] = df['bedroom'].replace(i, x[0])

df['bedroom'] = df['bedroom'].astype(int)
```

```python
df['bedroom'].unique()
```

```python
for i in df['bathroom']:
    x = i.split(' ')
    df['bathroom'] = df['bathroom'].replace(i, x[0])

df['bathroom'] = df['bathroom'].astype(int)
```

```python
df['bathroom'].unique()
```

```python
df.head()
```

```python
# if balcony is empty, then it is 0
df['balcony'] = df['balcony'].replace(np.nan, 0)
df['balcony'] = df['balcony'].astype(int)
```

```python
# if flooring is not 'Vitrified Tiles', 'Mosaic', 'Marble/Granite', 'Cement', 'Ceramic Tiles','Wooden' then make it empty

for i in df['flooring']:
    if i != 'Vitrified Tiles' and i != 'Mosaic' and i != 'Marble/Granite' and i != 'Cement' and i != 'Wooden':
        df['flooring'] = df['flooring'].replace(i, 0)


df['flooring'] = df['flooring'].replace('Vitrified Tiles', 1)
df['flooring'] = df['flooring'].replace('Mosaic', 2)
df['flooring'] = df['flooring'].replace('Marble/Granite', 3)
df['flooring'] = df['flooring'].replace('Cement', 4)
df['flooring'] = df['flooring'].replace('Wooden', 5)

df['flooring'] = df['flooring'].astype(int)

df['flooring'].unique()
```
