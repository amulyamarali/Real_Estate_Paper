```python
# read csv file 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read csv file
df = pd.read_csv('NO_BROKER_combined_file.csv')
df.head()
```

```python
# if parking is Yes or No then put floor value of the same index if parking is empty dont change
for i in range(len(df)):
    if df['parking'][i] == 'Yes' or df['parking'][i] == 'No' :
        print(df['parking'][i], "im in")
        if 'Bike' in str(df['floor'][i]) or 'Car' in str(df['floor'][i]):
            df['parking'][i] = df['floor'][i]
            print(df['parking'][i], "floor")
        if 'Bike' in str(df['facing'][i]) or 'Car' in str(df['facing'][i]):
            df['parking'][i] = df['facing'][i]
            print(df['parking'][i], "face")

# check if parking column has null values
df['parking'].isnull().sum()
```

```python
# if parking is No or Yes make it empty
for i in range(len(df)):
    if df['parking'][i] == 'Yes' or df['parking'][i] == 'No' :
        df['parking'][i] = np.nan
        
```

```python
# if "/" in parking make it empty
for i in range(len(df)):
    if "/" in str(df['parking'][i]):
        df['parking'][i] = np.nan
```

```python
# if parking is empty make it No
for i in range(len(df)):
    if pd.isnull(df['parking'][i]):
        df['parking'][i] = 'No'
```

```python
# print count of how many "/" in facing and floor and furnishing column
count = 0
for i in range(len(df)):
    if "/" in str(df['facing'][i]):
        count += 1
print(count)

count = 0
for i in range(len(df)):
    if "/" in str(df['floor'][i]):
        count += 1
print(count)

count = 0
for i in range(len(df)):
    if "/" in str(df['furnishing'][i]):
        count += 1
print(count)

```

```python
# if "/" in facing or furnishing and not in floor than transfer value of facing or furnishing to floor
for i in range(len(df)):
    if "/" in str(df['facing'][i]) and "/" not in str(df['floor'][i]):
        df['floor'][i] = df['facing'][i]
    if "/" in str(df['furnishing'][i]) and "/" not in str(df['floor'][i]):
        df['floor'][i] = df['furnishing'][i]

print(df['floor'].isnull().sum())
print(df['floor'].unique())
```

```python
# if North or South or East or West is not in  facing column and North or South or East or West is in furnishing column or remove column then transfer value to facing

for i in range(len(df)):
    if 'North' not in str(df['facing'][i]) and 'South' not in str(df['facing'][i]) and 'East' not in str(df['facing'][i]) and 'West' not in str(df['facing'][i]):
        if 'North' in str(df['furnishing'][i]) or 'South' in str(df['furnishing'][i]) or 'East' in str(df['furnishing'][i]) or 'West' in str(df['furnishing'][i]):
            df['facing'][i] = df['furnishing'][i]
        if 'North' in str(df['remove'][i]) or 'South' in str(df['remove'][i]) or 'East' in str(df['remove'][i]) or 'West' in str(df['remove'][i]):
            df['facing'][i] = df['remove'][i]


# print the unique values and their count in facing column
print(df['facing'].value_counts())
```

```python
df['furnishing'].unique()
```

```python
# if Semi, Fully furnished or Unfurnished is not in furnishing column and Semi, Fully furnished or Unfurnished is in remove column then transfer value to furnishing

for i in range(len(df)):
    if 'Semi' not in str(df['furnishing'][i]) and 'Fully' not in str(df['furnishing'][i]) and 'Unfurnished' not in str(df['furnishing'][i]):
        if 'Semi' in str(df['remove'][i]) or 'Fully' in str(df['remove'][i]) or 'Unfurnished' in str(df['remove'][i]):
            df['furnishing'][i] = df['remove'][i]

# print the unique values and their count in furnishing column
print(df['furnishing'].value_counts())
```
